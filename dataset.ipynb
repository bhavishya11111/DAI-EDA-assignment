### Education in Egypt

#### Description: This dataset contains information about 50,000 students in Egypt, including their personal details, parental education levels, type of education, and grades in various subjects. The data is synthetically generated to simulate real-world educational data and can be used for exploratory data analysis (EDA), classification tasks, and regression analysis.
Columns:

Student Name: The name of the student.
Student Age: The age of the student, ranging from 14 to 18 years.
Student Year: The year level of the student, ranging from Year 9 to Year 12.
Father Degree: The highest educational degree obtained by the student’s father (None, High School, Bachelor, Master, PhD).
Mother Degree: The highest educational degree obtained by the student’s mother (None, High School, Bachelor, Master, PhD).
Education Type: The type of education the student is enrolled in (IGCSE, IB, Thanweya).
Subject_1 to Subject_10: Grades for 10 different subjects, with values ranging from 20% to 100%, and a median around 75%-80%.

###  we take a 10% from the data to make EDA

```python

the_10_precentage = int(len(Data) * 0.1)
Data_10_precentage = Data[:the_10_precentage]
Data_10_precentage
```

### The code separates the Data_10_precentage DataFrame into numerical and categorical datasets for targeted analysis and visualization of their respective features.

```python
NumericaL_Data = Data_10_precentage.select_dtypes(include=[np.number])
Categorical_Data = Data_10_precentage.select_dtypes(include=[np.object_])
```

```python
print(f'The Nummerical Features Columns : \n {NumericaL_Data.columns} \n')
print(f'The Categorical Features Columns : \n  {Categorical_Data.columns}') 
```

```python
plt.figure(figsize=(18,8))
Data_10_precentage.hist(bins=50 ,
                        density=True ,
                       figsize=(18,10))
plt.title('The Distribution of Our DataSet')
plt.tight_layout()
plt.show()
```

```python
Categorical_Columns = ['Student year', 'Father Degree', 'Mother Degree', 'Education Type', 'Student Age']
for i in Categorical_Columns:
    x = Data_10_precentage[i].value_counts()
    plt.figure(figsize=(4,4))
    plt.pie(x=x.values ,
           labels=x.keys() ,
           explode=[0.01] * len(x.keys()),
           autopct='%1.1f%%',startangle= 90 ,
            wedgeprops={"edgecolor":"black"} ,
           shadow=True)
    plt.title(f'The {i} Distribution ')
    plt.tight_layout()
    plt.show()
```

#### The code generates pie charts for five categorical variables, illustrating student distributions by year, parental education, education type, and age, which helps identify demographic trends and supports future analyses of student performance.

```python
plt.figure(figsize=(18,8))
plt.subplot(2,2,1)
plt.title('The Father Degree Vs. Education Type')
sns.countplot(data=Data_10_precentage ,x='Father Degree' , hue='Education Type' , palette="plasma" )
plt.subplot(2,2,2)
plt.title('The Mother Degree Vs. Education Type')
sns.countplot(data=Data_10_precentage ,x='Mother Degree' , hue='Education Type' , palette="plasma" )
plt.tight_layout()
plt.show()
```

#### Conclusions from the dataset:
Effect of Education Type:

Students enrolled in IB (International Baccalaureate) performed slightly better on average (74.84) compared to students in IGCSE (74.68) and Thanweya (74.68). However, the differences in performance across the education types are minimal.
Student Year Performance:

The average scores across student years (Year 9 to Year 12) are very close:
Year 9: 74.78
Year 10: 74.77
Year 11: 74.70
Year 12: 74.69
This suggests that student performance remains consistent across different academic years.
Parental Education and Student Performance:

The correlation between father's education level and student performance is 0.015, indicating almost no correlation.
The correlation between mother's education level and student performance is -0.002, which also shows no meaningful relationship.
Summary:
There is no significant relationship between parental education and student performance.
The type of education (IB, IGCSE, Thanweya) shows only slight differences in performance.
Student performance remains consistent across academic years.
